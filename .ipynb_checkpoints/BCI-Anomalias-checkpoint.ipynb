```python
def _connect_mongo(db, host='localhost', port=27017, username=None, password=None):
    """ A util for making a connection to mongo """

    if username and password:
        mongo_uri = 'mongodb://%s:%s@%s:%s/%s' % (username, password, host, port, db)
        conn = MongoClient(mongo_uri)
    else:
        conn = MongoClient(host, port)


    return conn[db]
```

```python
def read_mongo(
       db, 
       collection, query=None, 
       host='localhost', port=27017, username=None, password=None,
       chunksize = 100, page_num=1, no_id=True):

    # Connect to MongoDB
    db = _connect_mongo(host=host, port=port, username=username, password=password, db=db)

    # Calculate number of documents to skip
    skips = chunksize * (page_num - 1)

    if not query:
        query = {}

    # Make a query to the specific DB and Collection
    cursor = db[collection].find(query).skip(skips).limit(chunksize)

    # Expand the cursor and construct the DataFrame
    df =  pd.DataFrame(list(cursor))

    # Delete the _id
    if no_id:
        del df['_id']

    return df
```

```python
df2 = df.drop("Currency_Cd", axis=1)
df2 = df2.drop("Acct_Balance_Dt", axis=1)
df2 = df2.drop("Acct_Banking_Group_Num", axis=1)
df2 = df2.drop("Account_Modifier_Num", axis=1)
df2 = df2.drop("Account_Num", axis=1)
df2 = df2.drop("GL_Acct_Num", axis=1)
```

```python
X = df2.values
n_samples, n_features = X.shape
```

```python
#----------------------------------------------------------------------
# Scale and visualize the embedding vectors
def plot_embedding(X, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    plt.figure(figsize=(20,15))
    ax = plt.subplot(111)
    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], str('+'),
                 #color=plt.cm.Set1(X_original[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)
```

```python
print("Computing PCA projection")
t0 = time()
std_scale = preprocessing.StandardScaler().fit(X)
X_train_std = std_scale.transform(X)

X_pca = decomposition.TruncatedSVD(n_components=2).fit_transform(X_train_std)
plot_embedding(X_pca,
               "Principal Components projection of the digits (time %.2fs)" %
               (time() - t0))
```

```python
#----------------------------------------------------------------------
# Isomap projection of the digits dataset
print("Computing Isomap embedding")
t0 = time()
n_neighbors = 30
std_scale = preprocessing.StandardScaler().fit(X)
X_train_std = std_scale.transform(X)

X_iso = manifold.Isomap(n_neighbors, n_components=2).fit_transform(X_train_std)
print("Done.")
plot_embedding(X_iso,
               "Isomap projection of the digits (time %.2fs)" %
               (time() - t0))
```
